```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("../data/teen_phone_addiction_dataset.csv")

data.info()
data.describe()
data.head()
```

Heatmap duoi day cho thay Addiction_Level co moi quan he chat che voi mot so bien quan trong. Cu the, Sleep_Hour(r=-0.22) co tuong quan am, cho thay thieu ngu anh huong den muc do nghien dien thoai (Addiction_Level). Phone_Checks_Per_Day(r=0.25), Time_On_Gaming(r=0.27), Time_on_Social_Media(r=0.31), va Apps_Used_Daily(r=0.32) co do tuong quan tuong doi voi muc do nghien dien thoai. Dac biet Daily_Usage_Hours co tuong quan duong manh nhat roi r = 0.6. Cac ket qua nay cung co cho gia thuyet nguoi thieu ngu, lien tuc su dung dien thoai hang ngay co xu huong co muc do nghien thoai.


```python
# Chi lay cot so
num_data = data.select_dtypes(include=["number"])

plt.figure(figsize = (10,8))
corr = num_data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap tuong quan giua cac bien so")
plt.show()
```

Bieu do boxplot cho thay thoi gian su dung dien thoai trung binh moi ngay (Daily_Usage_Hours) tap trung chu yeu trong khoang 4 - 6 gio, voi trung vi xap xi 5 gio / ngay. Tuy nhien, van ton tai mot so truong hop ngoai le (outliners) su dung tren 10 gio / ngay. Ket qua chi ra rang mac du phan lon hoc sinh co thoi gian su dung o muc trung binh cao, nhung van co mot nhom nho su dung dien thoai voi tan suat cuc ky lon, phan anh nguy co nghien nang.

```python
plt.figure(figsize = (8,5))
sns.boxplot(x=data["Daily_Usage_Hours"])
plt.title("Boxplot - Thoi gian trung binh su dung dien thoai")
plt.show()

```

- Cac bien hanh vi deu co phan phoi hop ly, nhung xuat hien nhung nhom outliners(vi du: > 10 gio dung dien thoai, > 100 lan kiem tra, > 18 ung dung / ngay)
